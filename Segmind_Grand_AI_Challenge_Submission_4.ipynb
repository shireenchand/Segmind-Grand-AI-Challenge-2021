```python
import pandas as pd
import sklearn
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('TRAIN.csv')
train_data
```

```python
train_data['StateHoliday'] = train_data['StateHoliday'].replace([0],'0')
```

```python
train_data = train_data.dropna()
```

```python
train_data.isna().sum()
```

```python
X = train_data.drop(['Sales','Customers','Date','Store'],axis=1)
Y = train_data['Sales']
Y
```

```python
store_data = pd.read_csv('store.csv')
store_data
```

```python
from sklearn.preprocessing import OneHotEncoder
object_cols = ["StoreType","Assortment"]
# Apply one-hot encoder to each column with categorical data
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(store_data[object_cols]))

# One-hot encoding removed index; put it back
OH_cols_train.index = store_data.index

# Remove categorical columns (will replace with one-hot encoding)
num_X_train = store_data.drop(object_cols, axis=1)

# Add one-hot encoded columns to numerical features
OH_X_train = pd.concat([num_X_train, OH_cols_train], axis=1)
```

```python
store_new_data = OH_X_train.copy()
store_new_data
```

```python
for i in range(0,7):
  X[i] = store_new_data[i]

X
```

```python
for i in range(0,7):
  X[i].fillna(X[i].mean(),inplace=True)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,Y_train,Y_valid = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
categorical_transformer = Pipeline(steps=[
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, ['DayOfWeek','StateHoliday'])
    ])
```

```python
from xgboost import XGBRegressor
my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05)
```

```python
my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                              ('my_model', my_model)
                             ])

my_pipeline.fit(X_train,Y_train)
```

```python
y_train_pred = my_pipeline.predict(X_train)
y_valid_pred = my_pipeline.predict(X_valid)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(Y_train,y_train_pred))

print("The model performance for training set")
print('RMSE is {}'.format(rmse))
print("\n")

rmse = np.sqrt(mean_squared_error(Y_valid,y_valid_pred))

print("The model performance for training set")
print('RMSE is {}'.format(rmse))
print("\n")
```
